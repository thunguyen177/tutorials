```python
!pip install sktime
import pandas as pd
import numpy as np
from sktime.datatypes._panel._convert import from_nested_to_multi_index
```

```python
# Example nested DataFrame
data = {
    "var1": [pd.Series([1, 2, 3]), pd.Series([4, 5, 6])],
    "var2": [pd.Series([7, 8, 9]), pd.Series([10, 11, 12])]
}
df_nested = pd.DataFrame(data)
df_nested
```

```python
# Convert to pd-multiindex format
df_multiindex = from_nested_to_multi_index(df_nested)

print(df_multiindex)
```

```python
from sktime.datatypes._panel._convert import from_nested_to_3d_numpy

df_3d = from_nested_to_3d_numpy(df_nested)
print(df_3d.shape)  # (2 instances, 3 timepoints, 2 features)
```

```python
df_3d
```

```python
from sktime.datatypes._panel._convert import from_3d_numpy_to_multi_index

df_multiindex = from_3d_numpy_to_multi_index(df_3d)
df_multiindex
```

```python
from sktime.datatypes._panel._convert import from_nested_to_long

# Convert nested DataFrame to a long DataFrame (standard tabular format)
df_long = from_nested_to_long(df_nested)
df_long
```
